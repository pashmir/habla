#### 1. Determine la longitud de la ventana que se debe aplicar a x(n) y la cantidad de puntos a usar en la DFT para diferenciar las dos frecuencias.

Aplicar una ventana hace a la señal x(n) equivale a convolucionar la señal con una sinc.
Para que ambas frecuencias sean distinguibles el ancho de los lóbulos principales debe ser tal que no formen un unico lóbulo. 

Voy a ver que pasa en 2 casos. El primero va ser cuando los lóbulos principales sean 0 en el punto donde se tocan. En base a los resultados voy a ver si es necesario reducir este ancho o, si es posible usar ventanas de anchos menores.

La transformada de fourier de una ventana de N longitud es

w(n) = 1(0<n<L) <=> W(w) = sin(w L *0.5))/sin(w/2) * e^j(w-L/2)

El lóbulo principal se hace 0 en w = 2 * pi / L

Busco w = 0.3 pi = 2 * pi / L  => 100 / 15 = L => 6.6 < L

Se puede ver que la longitud de la ventana necesaria es mucho menor que la cantidad de puntos necesarios para realizar la FFT y tener una resolución que permita ditinguir las frecuencias
Para poder distinguir las frecuencias es necesario que entre las frecuencias 0.4 pi y 0.8 pi haya al menos un punto. Con esto, el intervalo 0 a pi debería estar dividido en intervalos de al menos .2 pi. Con lo que NFFT deberia ser 10. Esto quiere decir que podria en teoria ventanear 7 muestras de x y rellenar con ceros para llegar a 10 y deberia poder diferenciar las senoidales.

```python
# Valores mínimos de L y NFFT
L = 7
NFFT = 10

# n, muestras de la señal continua
n = np.arange(L)

# x(n)
x = np.cos(2 * 100 / 500 * np.pi * n) + np.cos(2 * 200 / 500 * np.pi * n)
zeros = np.zeros(NFFT-L)
x = np.append(x,zeros)

# w, muestras de la DFT
w = np.arange(0,2,2/NFFT)

# X(w)
X = np.fft.fft(x)
Xabs = np.abs(X)

Xabs = Xabs/L # divido por L ya que es el valor medio de la ventana que usé
              # ME OLVIDE DE MULTIPLICAR POR PI, EL VALOR DE A Y B DEBERIA SER PI/2

# w(n): es una ventana rectangular
Win = np.fft.fft(np.append(np.zeros(L)+1,zeros))
absWin = np.abs(Win)

#Plot
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(w, Xabs)
ax1.set(xlabel='frec norm(w/pi)',
       title='|X(w)|')
ax1.grid()

ax2.plot(np.arange(0,2,2/absWin.size),absWin)
ax2.set(xlabel='frec norm(w/pi)',
       title='|W(w)|')
ax2.grid()
```

Sin embargo, con esta cantidad de puntos es solo es posible identificar la presencia de dos senoidales pero no precisar su frecuencia exacta. Dependiendo de la incerteza que se desee la ventana de la FFT deberá tener en frecuencia normalizada 2/incerteza_deseada puntos.
#### 2. Determinar la cantidad de puntos necesaria para identificar el valor de las constantes A y B

Acá lo que seria necesario lograr es obtener la menor "interferencia" entre sincs posibles. Esto se logra tomando más muestras de la señal x, es decir, tomando ventanas más grandes (L mayor).
El error en A va ser la altura de la sinc de B en la frecuencia de A. Entonces el error será el valor que tenga la ventana en f2-f1. Entonces:

e=sin(w L / 2) / sin (w / 2)|w=f2-f1 => e= sin((f2-f1) * L / 2) / sin ((f2-f1) /2)

e=sin( 0.4 pi * L / 2) / sin (0.2 pi)

Se puede ver entonces que es posible elegir un L tal que el error sea 0

0.4 * L = 2 * K => L = 5 * K

Cualquier multiplo de 5 tendrá los valores exactos de A y B EN ESTE CASO.

```python
# Valores mínimos de L y NFFT
L = 10
NFFT = 10

# n, muestras de la señal continua
n = np.arange(L)

# x(n)
x = np.cos(2 * 100 / 500 * np.pi * n) + np.cos(2 * 200 / 500 * np.pi * n)
zeros = np.zeros(NFFT-L)
x = np.append(x,zeros)

# w, muestras de la DFT
w = np.arange(0,2,2/NFFT)

# X(w)
X = np.fft.fft(x)
Xabs = np.abs(X)
Xabs = Xabs/L # divido por L ya que es el valor medio de la ventana que usé
              # ME OLVIDE DE MULTIPLICAR POR PI, EL VALOR DE A Y B DEBERIA SER PI/2
#Plot
fig, ax1 = plt.subplots()

ax1.plot(w, Xabs)
ax1.set(xlabel='frec norm(w/pi)',
       title='|X(w)|')
ax1.grid()
print("El valor de A es ",Xabs[2]," y el valor de B es ",Xabs[4])
```

#### 3. Con 8 veces mas puntos en la DFT (rellenar con 0) que resultados se obtienen? Y con una ventana de hanning?

```python
# Valores mínimos de L y NFFT
L_sinc = 7
L_hann = 14 # tomo el doble de muestras porque la ventana de hann tiene el doble
            # del ancho del lóbulo principal que la sinc
NFFT = 10 * 8

# n, muestras de la señal continua para la sinc
n = np.arange(L_sinc)

# x_sinc(n)
x_sinc = np.cos(2 * 100 / 500 * np.pi * n) + np.cos(2 * 200 / 500 * np.pi * n)
zeros = np.zeros(NFFT-L_sinc)
x_sinc = np.append(x_sinc,zeros)

# n, muestras de la señal continua para la hann
n = np.arange(L_hann)

#x_hann(n)
x_hann = np.cos(2 * 100 / 500 * np.pi * n) + np.cos(2 * 200 / 500 * np.pi * n)
win = np.hanning(L_hann)
x_hann = x_hann * win
zeros = np.zeros(NFFT-L_hann)
x_hann = np.append(x_hann,zeros)

# w, muestras de la DFT
w = np.arange(0,2,2/NFFT)

# X_sinc(w)
X_sinc = np.fft.fft(x_sinc)
Xabs_sinc = np.abs(X_sinc)
Xabs_sinc = Xabs_sinc/L # divido por L ya que es el valor medio de la ventana que usé
                        # ME OLVIDE DE MULTIPLICAR POR PI, EL VALOR DE A Y B DEBERIA SER PI/2

# X_hann(w)
X_hann = np.fft.fft(x_hann)
Xabs_hann = np.abs(X_hann)
# Nornmalizo
fact_escala = np.fft.fft(np.append(win,zeros))[0]
fact_escala = np.abs(fact_escala)
Xabs_hann = Xabs_hann/fact_escala # acá también faltaría multiplicar por pi

#Plot
fig, ax1 = plt.subplots()
ax1.plot(w, Xabs_sinc, w, Xabs_hann)
ax1.set(xlabel='frec norm(w/pi)',
       title='|X(w)|')
ax1.grid()
```

En este caso se usaron longitudes L que deban errores distintos de 0. Pero esto esta bueno para apreciar que la ventana de Hann a pesar de esto dio valores mucho mas cercanos a los correctos.
